```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=2)
```

```python
! pip install xgboost
import xgboost as xgb
```

```python
model=xgb.XGBRegressor()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_train)
```

```python
print(y_pred)
```

```python
from sklearn import metrics
score_1=metrics.r2_score(y_train,y_pred)
score_2=metrics.mean_absolute_error(y_train,y_pred)
print(f'r2 score {score_1}')
print(f"mean absolute error :{score_2}")
```

```python
y_pred1=model.predict(x_test)
```

```python
score_test=metrics.r2_score(y_test,y_pred1)
score_test2=metrics.mean_absolute_error(y_test,y_pred1)
print(f'r2 score {score_test}')
print(f"mean absolute error :{score_test2}")
```
